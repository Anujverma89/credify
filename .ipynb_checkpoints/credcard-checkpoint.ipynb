```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("dataset.csv")
# slicing the dataset:
data.drop(['customer_id', 'name', 'age', 'gender', 'owns_car', 'owns_house',
       'no_of_children','no_of_days_employed',
       'occupation_type', 'total_family_members', 'migrant_worker',
       'yearly_debt_payments','credit_limit_used(%)','prev_defaults', 'default_in_last_6months',
       'credit_card_default'],axis=1,inplace=True)
```

```python
name=['Yearly_income','Credit_limit','Credit_score']
data.columns=name
```

```python
#training over only 1000 rows
df=data[:1000]
```

```python
for index,row in df.iterrows():
    if(row.Yearly_income>300000):
        df.drop(index,inplace=True,axis=0)
        
for index,row in df.iterrows():
    if(row.Credit_limit>60000):
        df.drop(index,inplace=True,axis=0)
```

```python
#scatterplot of credit_limit vs Yearly_income 
x=df.Yearly_income
y=df.Credit_limit
plt.scatter(x,y)
plt.title("Credit_limit VS Yearly_income",fontsize=20)
plt.xlabel("Yearly_income")
plt.ylabel("Credit_limit")
plt.xticks(rotation=90)
plt.show()
```

```python
# Standardization and splitting of the data
x_values=df[["Yearly_income"]]
y_values=df[["Credit_limit"]]
#splitting traing and testing dataset
x_train,x_test,y_train,y_test=train_test_split(x_values,y_values,test_size=0.25,random_state=10)

#standardization of data
standardize=StandardScaler()
x_train=pd.DataFrame(standardize.fit_transform(x_train),columns=x_values.columns)

x_test=pd.DataFrame(standardize.transform(x_test),columns=x_values.columns)

y_train=pd.DataFrame(standardize.fit_transform(y_train),columns=y_values.columns)

y_test=pd.DataFrame(standardize.transform(y_test),columns=y_values.columns)
```

```python
regr=LinearRegression()
regr.fit(x_train,y_train)
# score=regr.score(x_train,y_train)
# print(score)
res=regr.predict(x_test)
print(res)
```

```python
plt.title('Loss / Mean Squared Error')
plt.plot(res, label='train')
plt.plot(y_test, label='test')
plt.legend()
plt.show()
```

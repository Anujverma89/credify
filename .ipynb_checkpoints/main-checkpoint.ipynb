```python
#importing dataprocessing modules
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# modules to build neural network
import tensorflow as tf
from tensorflow.keras import Model
from tensorflow.keras import Sequential
from tensorflow.keras.optimizers import Adam
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.losses import MeanSquaredLogarithmicError
from tensorflow.keras.layers import Dense, Dropout
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
#tosplit up the train and split data
from sklearn.model_selection import train_test_split

#importing gradio for UI deisng 
import gradio as gr

# incase to develop linear regressor 
```

```python
#impotring datafile and filtering it 
df=pd.read_csv("dataset.csv")
# df=df[0:5000]
df.drop(["customer_id","gender","name","owns_car","no_of_children","no_of_days_employed","occupation_type","total_family_members","migrant_worker","credit_limit_used(%)","prev_defaults","default_in_last_6months","credit_card_default"],axis=1,inplace=True)
names=["Age","Owns_house","Yearly_income","Debt_payment","Credit_limit","Credit_score"]
# names=["Age","Owns_house","Yearly_income","Debt_payment","Credit_limit","Credit_score"]
df.columns=names
encode={"Owns_house":{"Y":1,"N":0}}
df.replace(encode,inplace=True)
df.sort_values(by="Yearly_income",ascending=False,inplace=True)
for index,row in df.iterrows():
    if(row.Yearly_income>300000):
        df.drop(index,inplace=True,axis=0)
        
for index,row in df.iterrows():
    if(row.Debt_payment>116880):
        df.drop(index,inplace=True,axis=0)

for index,row in df.iterrows():
    if(row.Credit_limit>60000):
        df.drop(index,inplace=True,axis=0)
        
for index,row in df.iterrows():
    if(row.Debt_payment>62000):
        df.drop(index,inplace=True,axis=0)
```

```python
#Checking for erronious and missign data
df.dropna(axis=0,inplace=True)
df.describe()
# df.info()
```

```python
# Standardization and splitting of the data
# x_values=df[["Yearly_income","Credit_score"]]
x_values=df[["Age","Owns_house","Yearly_income","Debt_payment","Credit_score"]]
# x_values=df[["Yearly_income"]]
y_values=df[["Credit_limit"]]
#splitting traing and testing dataset
x_train,x_test,y_train,y_test=train_test_split(x_values,y_values,test_size=0.25,random_state=10)

#standardization of data
standardize=StandardScaler()
x_train=pd.DataFrame(standardize.fit_transform(x_train),columns=x_values.columns)

x_test=pd.DataFrame(standardize.transform(x_test),columns=x_values.columns)

y_train=pd.DataFrame(standardize.fit_transform(y_train),columns=y_values.columns)

y_test=pd.DataFrame(standardize.transform(y_test),columns=y_values.columns)
```

```python
# creating the model 
model=Sequential()

model.add(Dense(254,input_dim=5,activation='relu'))
model.add(Dense(128,activation='relu'))

# adding another layer
model.add(Dense(64,activation='relu'))

model.add(Dense(32,activation='relu'))
model.add(Dense(16,activation='relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(4,activation='relu'))
model.add(Dense(2,activation='relu'))
#adding outputlayer
model.add(Dense(1,activation='linear'))


#compiling the model

model.compile(optimizer=Adam(learning_rate=0.001),loss='mean_squared_error',metrics=['mse'])
#nodes in input layer is equal to the no of features in the model
model.summary()
```

```python
# training the model over here 
#length of data is 41021
history=model.fit(x_train,y_train,epochs=40,validation_data=(x_test,y_test),shuffle=True,batch_size=200)
```

```python
#evaluation part is only r
train_mse = model.evaluate(x_train, y_train, verbose=0)



plt.title('Loss / Mean Squared Error')
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.legend()
plt.show()
```
